```python
import numpy as np
from scipy.spatial import distance
import matplotlib.pyplot as plt
```

```python
def generate_random_data(dimensions, num_points=1000):
    # Generate random integer data points between 0 and 100 for given dimensions
    random_data = np.random.randint(0, 100, size=(num_points, dimensions), dtype=int)
    return random_data
```

```python
def calculate_euclidean_distances(data):
    # Calculate Euclidean distances between all pairs of data points
    distances = []
    num_points = data.shape[0]
    for i in range(num_points):
        for j in range(i+1, num_points):
            distances.append(distance.euclidean(data[i], data[j]))
    return distances
```

```python
def plot_histogram(data, title):
    # Plot a histogram of the input data
    plt.hist(data, bins='auto')
    plt.xlabel('Distance')
    plt.ylabel('Frequency')
    plt.title(title)
    plt.show()
```

```python
# Generate data sets of different dimensions
data_5 = generate_random_data(5)
data_2000 = generate_random_data(2000)
data_5000 = generate_random_data(5000)
data_10000 = generate_random_data(10000)
```

```python
# Calculate Euclidean distances for each data set
distances_5 = calculate_euclidean_distances(data_5)
distances_2000 = calculate_euclidean_distances(data_2000)
distances_5000 = calculate_euclidean_distances(data_5000)
distances_10000 = calculate_euclidean_distances(data_10000)
```

```python
# Plot histograms for each data set
plot_histogram(distances_5, 'Euclidean distances histogram (5 dimensions)')
plot_histogram(distances_2000, 'Euclidean distances histogram (2000 dimensions)')
plot_histogram(distances_5000, 'Euclidean distances histogram (5000 dimensions)')
plot_histogram(distances_10000, 'Euclidean distances histogram (10000 dimensions)')
```
